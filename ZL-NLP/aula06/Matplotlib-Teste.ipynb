```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
X = np.arange(-100, 100, 5)
X
```

```python
Y = X ** 3
Y
```

```python
plt.plot(X, Y)
```

```python
plt.plot(X, Y, marker="o", markersize=5, linestyle="dashed", color="#FFFF00")
```

```python
plt.scatter(X, Y, marker="o", color="#FFFF00")
plt.ylabel("Valor de X ao cubo")
plt.xlabel("Valor de X")
plt.text(0, 0, "Ponto 0")
plt.title( "Grafico de X" )
plt.savefig( "x3.png" )
plt.show()
```

```python
Y3 = X ** 3
Y2 = X ** 2
Y4 = X ** 4
```

```python
plt.plot(X, Y2, marker="o", color="red", label="X", markerfacecolor="pink", linewidth=3)
plt.plot(X, Y3, marker="o", color="blue", label="X", markerfacecolor="cyan")
plt.plot(X, Y4, marker="o", color="green", label="X4", markerfacecolor="lightgreen")
plt.ylabel("Valor de Y")
plt.xlabel("Valor de X")
plt.text(0, 0, "Ponto 0")
plt.title( "Grafico de X a potencia" )
plt.savefig( "x a potencias.png" )
plt.ylim(0, 1000)
plt.legend()
plt.axhline(600)
plt.axhline(200)
plt.axvline(-50)
plt.axvline(0)
plt.axvline(50)
plt.show()
```
