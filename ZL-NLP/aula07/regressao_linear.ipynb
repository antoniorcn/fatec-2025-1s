```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
X = np.array([150, 160, 155, 140, 170, 165, 152])
Y = np.array([300000, 350000, 320000, 280000, 400000, 370000, 310000])
```

```python
def funcao_linear(x, w, b):
    return (x * w) + b
```

```python
W = 2150
B = 0
Y_hat = []
custo_final = 0
for indice, x in enumerate(X):
    y_hat = funcao_linear( x, W, B )
    y_real = Y[indice]
    custo = abs((y_hat - y_real))
    custo_final += custo
    Y_hat.append( y_hat )
custo_final = custo_final / (2 * len(Y_hat))
custo_final
```

```python
Y_hat
```

```python
plt.scatter(X, Y)
plt.scatter(X, Y_hat, color="red")
plt.show()
```
