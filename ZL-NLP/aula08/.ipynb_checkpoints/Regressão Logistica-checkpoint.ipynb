```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
X = np.linspace(-10, 10, 100) 
```

```python
X_random = np.random.randint(-1000, 1000, 1000)
```

```python
def sigmoid(valor):
    return 1/(1 + np.exp(-valor))
```

```python
Y = sigmoid( X )
```

```python
Y_random = sigmoid( X_random )
```

```python
X
```

```python
plt.scatter(X_random, Y_random, color="blue")
plt.scatter(X, Y, color="red")
plt.show()
```
