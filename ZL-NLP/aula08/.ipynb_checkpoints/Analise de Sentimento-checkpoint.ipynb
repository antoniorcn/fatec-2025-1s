```python
import pandas as pd
```

```python
import sklearn as sk
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv("../../../dados/nlp/imdb-reviews-pt-br.csv", encoding="utf-8")
df
```

```python
vetorizador = CountVectorizer()
```

```python
bag_of_words_denso = vetorizador.fit_transform( df["text_pt"] )
```

```python
vocabulario = vetorizador.get_feature_names_out()
```

```python
bag_of_words = pd.DataFrame.sparse.from_spmatrix(  bag_of_words_denso, columns=vocabulario )
bag_of_words
```

```python
modelo = LogisticRegression()
```

```python
modelo.fit(bag_of_words, df["sentiment"])
```

```python
texto = "Este filme é muito ruim"
```

```python
vetorizador_previsao = CountVectorizer(vocabulary = vocabulario)
```

```python
bow_previsao_denso = vetorizador_previsao.fit_transform( [ texto ] )
```

```python
bow_previsao = pd.DataFrame.sparse.from_spmatrix( bow_previsao_denso, columns=vocabulario )
bow_previsao
```

```python
modelo.predict(bow_previsao)
```
