```python
import matplotlib.pyplot as plt
```

```python
features_X = [150, 160, 155, 140, 170, 165, 152]
output_Y = [300000, 350000, 320000, 280000, 400000, 370000, 310000]
```

```python
def calcular_custo( entradas_X : [], saidas_Y : [], w : float, b : float  ) -> float : 
    erros : float = 0
    for indice, entrada in enumerate(entradas_X): 
        y_hat : float = entrada * w + b
        y : float = saidas_Y[indice]
        erro : float = (y_hat - y) ** 2
        erros += erro
    return erros * 1 / (2 * len(entradas_X) )
```

```python
W = 10
B = 5
erro = calcular_custo( features_X, output_Y, W, B ) 
print(f"W: {W}\tB:{B}\tErro:{erro}")
```

```python
linha = []
for x in features_X:
    y = W * x + B
    linha.append(y)
linha
```

```python
plt.scatter( features_X, output_Y, color="blue" )
plt.plot( features_X, linha, color="red")
plt.show()
```
