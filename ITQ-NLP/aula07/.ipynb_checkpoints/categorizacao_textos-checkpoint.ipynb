```python
import pandas as pd
```

```python
import numpy as np
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
## Abrir o arquivo para leitura, carregando um dataframe do arquivo
```

```python
df = pd.read_csv( "../../../dados/nlp/news_sentiment_analysis.csv", encoding="utf-8")
```

```python
df['Description'] 
```

```python
vetorizador = TfidfVectorizer() 
```

```python
bow = vetorizador.fit_transform( df['Description'] ) 
```

```python
bow
```

```python
vocabulario = vetorizador.get_feature_names_out()
```

```python
bow_matrix = pd.DataFrame.sparse.from_spmatrix(bow, columns=vocabulario)
```

```python
bow_matrix.shape
```

```python
bow_matrix
```

```python
encoder = LabelEncoder()
```

```python
Y = encoder.fit_transform(df['Type'])
```

```python
model = LogisticRegression()
```

```python
model.fit(bow_matrix, Y) 
```

```python
texto = df["Description"][3120]
texto
```
