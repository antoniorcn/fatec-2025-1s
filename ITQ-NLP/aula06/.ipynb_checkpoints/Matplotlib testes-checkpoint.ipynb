```python
import matplotlib.pyplot as plt 
```

```python
import numpy as np
```

```python
X = np.arange(0, 10, 0.1)
X
```

```python
Y = X ** 2
Y
```

```python
plt.scatter(X, Y, marker='.')
```

```python
Y = np.sin(X)
Y
```

```python
C = np.cos(X)
```

```python
plt.plot(X, Y, "g+-", label = "Senos", markersize=10, linewidth=2)
plt.scatter(X, C, label = "Cosenos", color="green", s=0.1)
plt.ylabel("Seno de X")
plt.xlabel("Angulo em radianos")
plt.title("Seno dos Angulos")
plt.axhline( 0 )
plt.axvline( 7 )
plt.axvline( 3 )
plt.legend()
plt.savefig("./seno e coseno.jpg")
plt.show()
```
